# Use the data set to build a logistic regression model (using sklearn) and predict the class label. Divide the dataset into training and test set (70,30) using train_test_split method in sklearn

```python
dia.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(dia[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction','Age']], dia.Outcome, test_size = 0.3, random_state = 42)
```

```python
x_test
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)                   # Solve this ERROR !
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
model.score(x_test, y_test)
```

```python
model.predict_proba(x_test)

#model.predict_proba(x_test).round(4)
```

# Use the test data set and evaluate the performance using a confusion matrix. Visualize the confusion matrix using a heat map

```python
from sklearn.metrics import confusion_matrix
```

```python
con_mat = confusion_matrix(y_test, y_pred)
con_mat
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

print("True Positive: ", tp)
print("True Negative: ", tn)
print("False Positive: ", fp)
print("False Negative: ", fn)
```

```python
sb.heatmap(con_mat, annot = True, cmap = "RdBu", fmt = "0.0f", xticklabels = [0, 1], yticklabels = [0, 1])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")                                                # Binary Classification
```
