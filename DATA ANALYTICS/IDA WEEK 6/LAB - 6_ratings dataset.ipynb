# Q6. Find the 5 most similar user for user with user Id 10.

```python
import numpy as np
```

```python
allusers=rt_pivot.values
user10=allusers[0]
den1=np.sqrt(sum([np.square(x) for x in user10]))
cosinesim=[(10,1)]
i=1
for user in allusers[1:]:
    numer=[x*y for x,y in zip(user10,user)]
    den2=np.sqrt(sum([np.square(x) for x in user]))
    costheta = sum(numer)/(den1*den2)
    cosinesim.append((rt_pivot.index[i],costheta))
    i=i+1
cosinesim.sort(key=lambda x:x[1],reverse =True)
simusers=cosinesim[0:6]
simusers
```
