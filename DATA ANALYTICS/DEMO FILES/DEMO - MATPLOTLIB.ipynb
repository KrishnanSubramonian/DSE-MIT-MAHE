```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,label="Line-1")       # legend label is given as Line-1
plt.legend()
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,label="Line-1")       
plt.legend(loc="upper left")          # location of legend label is given as per choice
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,label="Line-1")       
plt.grid()                                # background is changed to grid 
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,label="Line-1")       
plt.grid(color="b")                                # background grid color changed to blue (b stands for blue)
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,label="Line-1")       
plt.grid(color="b",ls="-.")                        # background grid is made of '-' and '.' with linespace(ls) function 
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,label="Line-1")       
plt.grid(color="b",ls="-")                        # background grid is made of '-' with linespace(ls) function 
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,label="Line-1")       
plt.grid(color="b",ls="-.",lw="0.25")                # background grid is made lighter with linewidth(lw) function 
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,"r",label="Line-1")                 # the line of graph changed to red      
plt.grid(color="b",ls="-.",lw="0.25")                
plt.show() 
```

```python
x = [1,2,1]
y = [2,4,6]
plt.plot(x,y,"r",lw="5",label="Line-1")          # the line of graph is made thick with linewidth(lw) function   
plt.grid(color="b",ls="-.",lw="0.25")                
plt.show() 
```

```python
countries = pd.read_csv("countries.csv")
countries
```

```python
afghan = countries[countries.country=="Afghanistan"]
afghan
```

```python
us = countries[countries.country=="United States"]
us
```

```python
china = countries[countries.country=="China"]
china
```

```python
plt.plot(us.year,us.population)         # year to population of a country 
plt.show()                              # Note that the y-axis is in millions 
```

```python
plt.plot(us.year,us.population/10**6)         # Y-Axis grid from 1 million to 1 unit  
plt.show()
```

```python
plt.plot(us.year,us.population/10**6)   
plt.plot(china.year,china.population/10**6)           # comparission of US and CHINA population Vs year 
plt.show()
```

```python
plt.plot(us.year,us.population/10**6)   
plt.plot(china.year,china.population/10**6)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Compare") 
plt.legend(["United States","china"])                                   # comparing in line plot 
plt.grid()
plt.show()
```

```python
plt.scatter(us.year,us.population/10**6)   
plt.scatter(china.year,china.population/10**6)
plt.xlabel("Year")
plt.ylabel("Population")                                     # comparing in scatter plot 
plt.title("Compare") 
plt.legend(["United States","china"])
plt.grid()
plt.show()
```
