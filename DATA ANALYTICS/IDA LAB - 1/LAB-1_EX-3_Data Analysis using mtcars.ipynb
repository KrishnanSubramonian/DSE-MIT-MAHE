```python
# Lab - 1: Statistical Analysis using Python: -

# EXERCISE - 3 – Data Analysis using mtcars 

import pandas as pd
mtcars=pd.read_csv("mtcars dataset csv.csv")
```

```python
# Meta data: - It's a data frame with 32 observations on 11 (numeric) variables where mpg is Miles/(US) gallon, 
#             cyl is Number of cylinders, disp is Displacement, hp is Gross horsepower, drat is Rear axle ratio,
#             wt is Weight ( in 1000 lbs), qsec is 1/4 mile time, vs is Engine (0 = V-shaped, 1 = straight)
#             am is Transmission (0 = automatic, 1 = manual), gear is Number of forward gears, carb is Number of carburetors

# This data was extracted from the 1974 Motor Trend US magazine, and comprises fuel consumption and 10 aspects of automobile design and performance for 32 automobiles (1973–74 models).

mtcars
```

```python
# Find the car with the best mpg

MaxMpg=mtcars["mpg"].max()
mtcars[mtcars["mpg"] == MaxMpg]
```

```python
# Find the car with the worst mpg

mtcars[mtcars["mpg"] == mtcars["mpg"].min()]
```

```python
# Find the car with the best horsepower

MaxHp=mtcars["hp"].max()
mtcars[mtcars["hp"]==MaxHp]
```

```python
# Find 5 number summary of displacement 

mtcars["disp"].describe()
```

```python
# Find median horse power

mtcars["hp"].median()
```

```python
# What is average mpg for manual vs. automatic cars

             # If the attribute 'am' is '0' then the car is manual

             # If the attribute 'am' is '1' then the car is automatic

mtcars.groupby('am')["mpg"].mean()
```

```python
# Draw a histogram of miles per gallon

mtcars.hist(column='mpg')
```

```python
# Boxplot of mpg for each cylinder type

mtcars.boxplot(column = 'mpg', by = 'cyl')


  #  mtcars["cyl"].plot.box()
```

```python
# Create a crosstab displaying count of automatic vs. manual cars

mtcars.am.value_counts()
```

```python
# Create a crosstab displaying count of “am vs cyl”

pd.crosstab(mtcars["am"],mtcars["cyl"])
```

```python
# What is the correlation between the weight of the car and mpg

mtcars['mpg'].corr(mtcars['wt'])          


# For the obtaining the correlation table of all attributes use: - mtcars.corr()


# There is a High degree of correlation between the variables weight of the car and mpg of the car 

#            The Negative sign here tells us that as mpg of a car increases the weight of the car decreases
#                                                  OR
#                   The attributes 'mpg' and 'wt' are inversely proportional to each other
```
